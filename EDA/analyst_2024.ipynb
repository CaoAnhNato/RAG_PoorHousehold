```python
import pandas as pd
import glob
import os
import random
```

```python
# Load mỗi file .xlsx ở folder '2024/' vào 1 dataframe riêng (lưu trong dictionary)
path_2024 = '../2024'
all_files = glob.glob(os.path.join(path_2024, "*.xlsx"))

dataframes_2024 = {}
for filename in all_files:
    # Lấy tên file (không có đuôi .xlsx) làm tên dataframe trong dict
    file_name = os.path.basename(filename).split('.')[0]
    dataframes_2024[file_name] = pd.read_excel(filename)

print("Danh sách các dataframe (Huyện) đã load:", list(dataframes_2024.keys()))
```

```python
def get_PoorFam_by_Period(df):
    """
    Lọc ra các rows có cùng giá trị cột 'family.hostName' nhưng khác giá trị 'classify'
    và hiển thị ngẫu nhiên 1 trường hợp.
    """
    # Tìm các gia đình có nhiều hơn 1 giá trị 'classify'
    grouped = df.groupby('family.hostName')['classify'].nunique()
    target_families = grouped[grouped > 1].index.tolist()
    
    if not target_families:
        print("Không tìm thấy trường hợp nào có cùng 'family.hostName' nhưng khác 'classify'.")
        return None
    
    # Chọn ngẫu nhiên 1 gia đình
    random_family = random.choice(target_families)
    
    # Lấy toàn bộ các dòng của gia đình đó
    result = df[df['family.hostName'] == random_family]
    
    return result
```

```python
# Kiểm tra ngẫu nhiên 1 trường hợp ở Huyện Cư Jút
# Bạn có thể thay 'Huyện Cư Jút' bằng các tên huyện khác trong dataframes_2024
target_df = dataframes_2024['Huyện Cư Jút']
random_case = get_PoorFam_by_Period(target_df)

if random_case is not None:
    display(random_case)
```
