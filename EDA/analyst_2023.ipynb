# Đọc dữ liệu từ thư mục 2023
Tiến hành đọc tất cả các tệp `.xlsx` trong thư mục `2023/` bằng thư viện `pandas`.

```python
import pandas as pd
import glob
import os

# Đường dẫn đến thư mục 2023 (từ thư mục EDA/)
folder_path = os.path.join('..', '2023')

# Đọc từng file vào các biến riêng biệt
df_cu_jut = pd.read_excel(os.path.join(folder_path, 'Huyện Cư Jút.xlsx'))
df_krong_no = pd.read_excel(os.path.join(folder_path, 'Huyện Krông Nô.xlsx'))
df_tuy_duc = pd.read_excel(os.path.join(folder_path, 'Huyện Tuy Đức.xlsx'))
df_dak_glong = pd.read_excel(os.path.join(folder_path, 'Huyện Đăk Glong.xlsx'))
df_dak_mil = pd.read_excel(os.path.join(folder_path, 'Huyện Đắk Mil.xlsx'))
df_dak_rlap = pd.read_excel(os.path.join(folder_path, 'Huyện Đắk RLấp.xlsx'))
df_dak_song = pd.read_excel(os.path.join(folder_path, 'Huyện Đắk Song.xlsx'))
df_gia_nghia = pd.read_excel(os.path.join(folder_path, 'Thành phố Gia Nghĩa.xlsx'))

# Kiểm tra thử dữ liệu của một huyện
print("Dữ liệu Huyện Cư Jút:")
print(df_cu_jut.head())
```

```python
df_cu_jut.info()
```

```python
# Lọc các dòng có 'family.code' và 'family.hostName' bị lặp trong df_cu_jut
duplicates = df_cu_jut[df_cu_jut.duplicated(subset=['family.code', 'family.hostName'], keep=False)]

print(f"Số lượng dòng trùng lặp (Mã hộ & Tên chủ hộ): {len(duplicates)}")
duplicates.sort_values(by=['family.code', 'family.hostName'])
```

```python
# Hiển thị các dòng có 'family.hostGender' là null trong df_cu_jut
null_gender = df_cu_jut[df_cu_jut['family.hostGender'].isna()]

print(f"Số lượng dòng có giới tính chủ hộ bị thiếu: {len(null_gender)}")
null_gender.head(20) # Hiển thị 20 dòng đầu tiên bị thiếu
```

## Kiểm tra dữ liệu các huyện còn lại
Tiến hành kiểm tra trùng lặp mã hộ (`family.code`) và thiếu thông tin giới tính (`family.hostGender`) cho các huyện khác.

```python
# Huyện Krông Nô
print(f"--- Huyện Krông Nô ---")
dup_krong_no = df_krong_no[df_krong_no.duplicated(subset=['family.code', 'family.hostName'], keep=False)]
null_gen_krong_no = df_krong_no[df_krong_no['family.hostGender'].isna()]

print(f"Số lượng trùng (Mã hộ & Tên chủ hộ): {len(dup_krong_no)}")
print(f"Số lượng thiếu giới tính: {len(null_gen_krong_no)}")
if len(null_gen_krong_no) > 0:
    display(null_gen_krong_no)
```

```python
# Huyện Tuy Đức
print(f"--- Huyện Tuy Đức ---")
dup_tuy_duc = df_tuy_duc[df_tuy_duc.duplicated(subset=['family.code', 'family.hostName'], keep=False)]
null_gen_tuy_duc = df_tuy_duc[df_tuy_duc['family.hostGender'].isna()]

print(f"Số lượng trùng (Mã hộ & Tên chủ hộ): {len(dup_tuy_duc)}")
print(f"Số lượng thiếu giới tính: {len(null_gen_tuy_duc)}")
if len(null_gen_tuy_duc) > 0:
    display(null_gen_tuy_duc)
```

```python
# Huyện Đăk Glong
print(f"--- Huyện Đăk Glong ---")
dup_dak_glong = df_dak_glong[df_dak_glong.duplicated(subset=['family.code', 'family.hostName'], keep=False)]
null_gen_dak_glong = df_dak_glong[df_dak_glong['family.hostGender'].isna()]

print(f"Số lượng trùng (Mã hộ & Tên chủ hộ): {len(dup_dak_glong)}")
print(f"Số lượng thiếu giới tính: {len(null_gen_dak_glong)}")
if len(null_gen_dak_glong) > 0:
    display(null_gen_dak_glong)
```

```python
# Huyện Đắk Mil
print(f"--- Huyện Đắk Mil ---")
dup_dak_mil = df_dak_mil[df_dak_mil.duplicated(subset=['family.code', 'family.hostName'], keep=False)]
null_gen_dak_mil = df_dak_mil[df_dak_mil['family.hostGender'].isna()]

print(f"Số lượng trùng (Mã hộ & Tên chủ hộ): {len(dup_dak_mil)}")
print(f"Số lượng thiếu giới tính: {len(null_gen_dak_mil)}")
if len(null_gen_dak_mil) > 0:
    display(null_gen_dak_mil)
```

## Tính tỷ lệ % và hiển thị các dòng trùng lặp mã hộ (`family.code`)
Phần này tính toán tỷ lệ phần trăm các dòng bị lặp dựa trên mã hộ và hiển thị các dòng đó.

```python
# Huyện Cư Jút
dups = df_cu_jut[df_cu_jut.duplicated(subset=['family.code', 'family.hostName'], keep=False)]
percent = (len(dups) / len(df_cu_jut)) * 100
print(f"Huyện Cư Jút: {percent:.2f}% dòng bị lặp ({len(dups)}/{len(df_cu_jut)})")
display(dups.sort_values(['family.code', 'family.hostName']).head(10))
```
